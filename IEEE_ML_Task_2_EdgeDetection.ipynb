```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
img = cv.imread('/content/car1.webp')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
blurred = cv.GaussianBlur(gray, (9, 9), 0)
```

```python
sobel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])  # Gradient in x-direction
sobel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])  # Gradient in y-direction
```

```python
grad_x = cv.filter2D(blurred, -1, sobel_x)  # Gradient in x-direction
grad_y = cv.filter2D(blurred, -1, sobel_y)  # Gradient in y-direction
```

```python
magnitude = np.sqrt(grad_x**2 + grad_y**2)
magnitude = np.uint8(magnitude)  # Convert to 8-bit image for display
```

```python
magnitude = cv.normalize(magnitude, None, 0, 255, cv.NORM_MINMAX)
```

```python
plt.subplot(131)
plt.imshow(gray, cmap='gray')
plt.title('Grayscale Image')
plt.xticks([]), plt.yticks([])

plt.subplot(132)
plt.imshow(blurred, cmap='gray')
plt.title('Blurred Image')
plt.xticks([]), plt.yticks([])

plt.subplot(133)
plt.imshow(magnitude, cmap='gray')
plt.title('Edge Magnitude')
plt.xticks([]), plt.yticks([])

# Display the plot
plt.show()
```
